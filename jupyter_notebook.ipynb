## Import everything here

```python
from tog.models.kg import Neo4jKnowledgeGraph
from tog.pipeline import GroqEntityExtractor
from tog.pipeline import EntityMapper
from tog.pipeline import Neo4jMappingHandler
from tog.pipeline import ExplorationLoop


from pprint import pprint
from dotenv import load_dotenv

load_dotenv()
```

### Entity Extractor Usage

```python
# Example usage
llm_extractor = GroqEntityExtractor("llama-3.3-70b-versatile")
sample_text = """what percentage of 1500 participants using medical cannabis to treat anxiety."""

entities = llm_extractor.extract_entities(sample_text)
print("-" * 50)
pprint(entities)
print(len(entities))
```

### Entity Mapper Usage

```python
kg = Neo4jKnowledgeGraph()
mapping_handler = Neo4jMappingHandler(kg=kg)

entity_mapper = EntityMapper(kg, mapping_handler)

# Example extracted entities
extracted_entities = ['medical cannabis', 'anxiety', 'participants', 'percentage', '1500']

mapped_entities = entity_mapper.map_entities(extracted_entities)

print("-" * 50)
pprint(f"{len(mapped_entities)} Mapped Entities:")
pprint("Mapped Entities:")
# for entity in mapped_entities:
#     pprint(f"ID: {entity.id}, Name: {entity.name}, Type: {entity.type}")
#     pprint(f"Metadata: {entity.metadata}")
pprint(mapped_entities)
```
